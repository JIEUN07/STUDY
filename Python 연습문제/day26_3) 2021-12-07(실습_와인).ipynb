### 결과가 좋지않아 다른 모델로 예측

```python
fixed_r = np.array(red["fixed acidity"])
print(fixed_r)
free_r = np.array(red["free sulfur dioxide"])
print(free_r)
residual_r = np.array(red["residual sugar"])
print(residual_r)
sulphates_r = np.array(red["sulphates"])
print(sulphates_r)
alcohol_r = np.array(red["alcohol"])
print(alcohol_r)
quality_r = np.array(red["quality"])
print(quality_r)
```

```python
# 기울기 절편초기화
a1_r = 0
a2_r = 0
a3_r = 0
a4_r = 0
a5_r = 0
b_r = 0

# 학습률 0.001
# 반복횟수 10000
lr = 0.001
epochs = 10000
```

```python
for i in range(epochs):
    pre_q = (a1_r*fixed_r) + (a2_r*free_r) + (a3_r*residual_r) + (a4_r*sulphates_r) + (a5_r*alcohol_r) +b_r
    err = quality_r - pre_q
    
    diff_a1 = -(2/len(fixed_r)) * sum(err*fixed_r)
    diff_a2 = -(2/len(free_r)) * sum(err*free_r)
    diff_a3 = -(2/len(residual_r)) * sum(err*residual_r)
    diff_a4 = -(2/len(sulphates_r)) * sum(err*sulphates_r)
    diff_a5 = -(2/len(alcohol_r)) * sum(err*alcohol_r)
    diff_b_r = -(2/len(quality_r)) * sum(err) 
    
    a1_r -= lr * diff_a1
    a2_r -= lr * diff_a2
    a3_r -= lr * diff_a3
    a4_r -= lr * diff_a4
    a5_r -= lr * diff_a5
    b_r -= lr * diff_b_r
    print(a1_r, a2_r, a3_r, a4_r, a5_r, b_r)
    
    if i % 100 == 0:
        print('Epoch = %.04d, a1 = %.04f, a2 = %.04f, a3 = %.04f, a4 = %.04f, a5 = %.04f, b = %.04f' %(i, a1_r, a2_r, a3_r, a4_r, a5_r, b_r))
```

```python
print('y = %.04fx1 + %.04fx2 + %.04fx3 + %.04fx4 + %.04fx5 + %.04f' %(a1_r, a2_r, a3_r, a4_r, a5_r, b_r))
```

```python
pre_q_r_1 = []
for i in range(len(quality_r)):
    pre_q_r_1.append(round(0.0876*fixed_r[i] + 0.0023*free_r[i] -0.0133*residual_r[i] + 0.4148*sulphates_r[i] + 0.4263*alcohol_r[i] + 0.1823))

print(pre_q_r_1)
```

```python
(quality_r == pre_q_r_1).sum()
```

```python
len(pre_q_r_1)
```

```python
# 정확도
(quality_r == pre_q_r_1).sum() / len(pre_q_r_1)
```
