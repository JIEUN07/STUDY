# titanic 분석


```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.set(context='notebook', 
        style='darkgrid', 
        palette='deep', 
        # font='sans-serif', 
        font_scale=1, 
        color_codes=True)
```

성별

```python
sns.histplot(titanic, x = "sex", shrink=0.8)
```

```python
sns.histplot(titanic, x = "sex", hue = "survived", multiple="dodge", shrink=0.8)
```

```python
sns.histplot(titanic, x = "sex", hue = "survived", multiple = "fill",shrink=0.8)
plt.show()
```

객실 등급

```python
sns.histplot(titanic, x = "pclass")
```

```python
sns.histplot(titanic, x = "pclass", hue = "survived", multiple="dodge")
plt.show()
```

```python
sns.histplot(titanic, x = "pclass", hue = "survived", multiple = "fill")
plt.show()
```

```python
sns.catplot(data = titanic, x = "pclass", col="sex", hue = "survived", kind = "count")
plt.show()
```

요금

```python
sns.histplot(titanic, x = "fare", element ="poly")
plt.show()
```

```python
not_zero_fare = titanic.query("fare>0")
sns.histplot(not_zero_fare, x = "fare", element ="poly", hue="survived", log_scale=True)
plt.show()
```
